```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from src.ingestion.download import load_application_train
from src.config import RAW_DIR

df = load_application_train()
print(df.shape)
df.head()
```

```python
# Overall default rate
print(f"Default rate: {df['TARGET'].mean():.2%}")
print(f"\nTarget distribution:")
print(df['TARGET'].value_counts())
```

```python
# Income and credit distribution
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
df['AMT_INCOME_TOTAL'].clip(0, 500_000).hist(ax=axes[0], bins=50, color='steelblue')
axes[0].set_title('Income Distribution')
df['AMT_CREDIT'].clip(0, 1_500_000).hist(ax=axes[1], bins=50, color='coral')
axes[1].set_title('Credit Amount Distribution')
plt.tight_layout()
plt.show()
```

```python
# Correlation of ext_sources with target
ext_cols = ['EXT_SOURCE_1', 'EXT_SOURCE_2', 'EXT_SOURCE_3']
corr = df[ext_cols + ['TARGET']].corr()
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Correlation: External Scores × Target')
plt.show()
```
