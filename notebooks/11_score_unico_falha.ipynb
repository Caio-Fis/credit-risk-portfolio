# Single Score — Quantitative Demonstration of Limitations

The single score is not wrong — it is insufficient.

```python
import sys; sys.path.insert(0, '..')
```

```python
from src.contextual.data_generator import generate_dataset
from src.contextual.interaction_model import compare_single_vs_contextual

df = generate_dataset(n=5000, seed=42)
comparison = compare_single_vs_contextual(df, n_examples=5)
comparison
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Visualize EL difference for the same clients in different contexts
sample = comparison.head(30)
fig, ax = plt.subplots(figsize=(12, 5))
x = range(len(sample))
ax.bar(x, sample['el_single'], label='EL Single Score', alpha=0.7, color='coral')
ax.bar(x, sample['el_contextual'], label='EL Contextual Score', alpha=0.7, color='steelblue')
ax.set_ylabel('Expected Loss')
ax.set_title('Single Score vs. Contextual Score — same clients, different contexts')
ax.legend()
plt.tight_layout()
plt.show()
```

```python
# Quantitative conclusion
delta = comparison['el_delta'].abs()
print(f"Mean EL difference between models: {delta.mean():,.2f}")
print(f"Maximum EL difference: {delta.max():,.2f}")
print(f"\nThe single score underestimates risk for long tenors and overestimates for short ones.")
print("Using a single score means mispricing the majority of contracts.")
```
