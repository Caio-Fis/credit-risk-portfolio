O score único não é errado — é insuficiente.

```python
from src.contextual.data_generator import generate_dataset
from src.contextual.interaction_model import compare_single_vs_contextual

df = generate_dataset(n=5000, seed=42)
comparison = compare_single_vs_contextual(df, n_examples=5)
comparison
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Visualiza diferença de EL para os mesmos clientes em diferentes contextos
sample = comparison.head(30)
fig, ax = plt.subplots(figsize=(12, 5))
x = range(len(sample))
ax.bar(x, sample['el_single'], label='EL Score Único', alpha=0.7, color='coral')
ax.bar(x, sample['el_contextual'], label='EL Score Contextual', alpha=0.7, color='steelblue')
ax.set_ylabel('Expected Loss (R$)')
ax.set_title('Score Único vs. Score Contextual — mesmos clientes, contextos diferentes')
ax.legend()
plt.tight_layout()
plt.show()
```

```python
# Conclusão quantitativa
delta = comparison['el_delta'].abs()
print(f"Diferença média de EL entre modelos: R$ {delta.mean():,.2f}")
print(f"Diferença máxima de EL: R$ {delta.max():,.2f}")
print(f"\nO score único subestima o risco em prazos longos e superestima em curtos.")
print("Usar score único implica precificar incorretamente a maioria dos contratos.")
```
