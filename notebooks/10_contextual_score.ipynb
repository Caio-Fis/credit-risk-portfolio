```python
from src.contextual.data_generator import generate_dataset
from src.contextual.interaction_model import train_contextual, score_by_context

df = generate_dataset(n=5000, seed=42)
model_ctx, feature_names = train_contextual(df)
```

```python
# Specific client profile
client_profile = {
    'score_financeiro': 0.5,
    'idade_empresa_anos': 5,
    'faturamento_anual': 500_000,
    'has_collateral': 0,
    'ead': 50_000,
}

scores = score_by_context(client_profile, model_ctx, feature_names)
scores
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

pivot = scores.pivot(index='tenor_months', columns='product_type', values='pd_contextual')
sns.heatmap(pivot, annot=True, fmt='.3f', cmap='RdYlGn_r')
plt.title('Contextual PD: Tenor × Product (same client)')
plt.show()
```
