```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from src.ingestion.download import load_application_train
from src.config import RAW_DIR

df = load_application_train()
print(df.shape)
df.head()
```

```python
# Taxa de default geral
print(f"Taxa de default: {df['TARGET'].mean():.2%}")
print(f"\nDistribuição do target:")
print(df['TARGET'].value_counts())
```

```python
# Distribuição de renda e crédito
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
df['AMT_INCOME_TOTAL'].clip(0, 500_000).hist(ax=axes[0], bins=50, color='steelblue')
axes[0].set_title('Distribuição de Renda')
df['AMT_CREDIT'].clip(0, 1_500_000).hist(ax=axes[1], bins=50, color='coral')
axes[1].set_title('Distribuição do Crédito Concedido')
plt.tight_layout()
plt.show()
```

```python
# Correlação dos ext_sources com o target
ext_cols = ['EXT_SOURCE_1', 'EXT_SOURCE_2', 'EXT_SOURCE_3']
corr = df[ext_cols + ['TARGET']].corr()
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Correlação: Scores Externos × Target')
plt.show()
```
